### Ques 1.	Implement crawlers which take as input a url, a search word and maximum number of pages to be searched and returns as output all the web pages it searched till it found the search word on a web page or return failure.

## Breadth-First-Search

```python
# Importing libraries
from bs4 import BeautifulSoup
from urllib.request import urljoin
from urllib.request import urlparse
import requests
```

```python
# Urls with same domains
links_intern = set()

# Urls with different domains
links_extern = set()

input_url = input("Enter the url: ")
depth = int(input("Enter the depth level of the search: "))
```

```python
# Method for crawling a url at next level
def level_crawler(input_url):
    temp_urls=set()
    # Parsing the entered URL
    current_url_domain=urlparse(input_url).netloc
    
    # Creates beautifulsoup object to extract html tags
    beautiful_soup_object = BeautifulSoup(requests.get(input_url).content, "lxml")
    
    # Access all anchor tags from input url page and divide them
    # into internal and externalcategories
    for anchor in beautiful_soup_object.findAll("a"):
        href = anchor.attrs.get("href")
        if(href!="" or href!=None):
            href=urljoin(input_url, href)
            href_parsed=urlparse(href)
            href=href_parsed.scheme
            href+="://"
            href+=href_parsed.netloc
            href+=href_parsed.path
            final_parsed_href=urlparse(href)
            is_valid=bool(final_parsed_href.scheme) and bool(
                final_parsed_href.netloc)
            if is_valid:
                if current_url_domain not in href and href not in links_extern:
                    # Not from entered link
                    print("External - {}".format(href))
                    links_extern.add(href)
                if current_url_domain in href and href not in links_intern:
                    # From entered link
                    print("Internal - {}".format(href))
                    links_intern.add(href)
                    temp_urls.add(href)
    return temp_urls
```

```python
if(depth == 0):
    print("Intern - {}".format(input_url))
elif(depth == 1):
    level_crawler(input_url)
else:
    queue = []
    queue.append(input_url)
    for j in range(depth):
        for count in range(len(queue)):
            url = queue.pop(0)
            urls = level_crawler(url)
            for i in urls:
                queue.append(i)
```



## Depth-First-Search

```python
def search_dfs(base, path,visited,max_depth=3, depth=0):
    if depth < max_depth:
        try:
            soup = BeautifulSoup(requests.get(base + path).text, "html.parser")
            for link in soup.find_all("a"):
                href = link.get("href")
                if href not in visited:
                    visited.add(href)
                    print(f"\tDepth {depth}: {href}")
                    if href.startswith("http"):
                        search_dfs(href, "", visited,max_depth-1, depth + 1)
                    else:
                        search_dfs(base, href,visited,max_depth-1, depth + 1)
        except:
            pass
```

```python
link = input("Enter the url: ")
```

```python
search_dfs(link, "", set([link]))
```
